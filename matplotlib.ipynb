```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,20)
y1 = np.sin(x**2)
y2 = np.cos(x)
y3 = np.tanh(x)
y4 = np.tanh(x**3)
y5 = x**2
y6 = np.log(x+1)

# Method 1: Using plt.plot() to show with current figure and axes
plt.figure(1) # This is optional as plot autogenerates figure and axes; Added here for clarity, can be commented
plt.plot(x, y1)
plt.plot(x, y1**2) # adds graph to the same figure

plt.figure(2) # This is added to create new figure and not add graph to same figure above
plt.plot(x, y2)

# Method 2: Create a figure with multiple axes to plot various graphs
plt.figure(3) # Again not required explicitely as plt.subplots creates a new figure by itself

f, ax = plt.subplots(2, 2) # This creates 4 subplots in same figure
ax[0, 0].plot(x, y3)
ax[0, 1].plot(x, y4)
ax[1, 0].plot(x, y5)
ax[1, 1].plot(x, y6)
plt.show() # This is optional because ipython is by default in intercative mode and shows all figures; Added for clarity, can be commented

# Show an image with plt.imshow()
from scipy import misc
img = misc.ascent() # Random sample image for experimenting
plt.figure(4)
plt.imshow(img)
```
